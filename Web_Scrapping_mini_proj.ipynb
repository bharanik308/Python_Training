```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.carwale.com/audi-cars/")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
soup
```

```python
all=soup.find_all("div",{"class":"o-C o-cF o-c9 o-bT o-cp o-f"})
all
```

```python
for item in all:
    
    print(item.find_all("h3",{"class":"o-j4 o-jq o-hM o-c4"})[0].text)
    print(item.find_all("span",{"class":"o-jJ o-jr o-j2 o-kJ"})[0].text)
    print(item.find_all("div",{"class":"o-f o-j0 o-jf o-bO o-cV o-kJ o-jL"})[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["car_name"]=item.find_all("h3",{"class":"o-j4 o-jq o-hM o-c4"})[0].text
    d["car_price"]=item.find_all("span",{"class":"o-jJ o-jr o-j2 o-kJ"})[0].text
    d["car_description"]=item.find_all("div",{"class":"o-f o-j0 o-jf o-bO o-cV o-kJ o-jL"})[0].text
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
